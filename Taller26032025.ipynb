```python
# quiero generar una matriz con datos aleatorios con los siguientes campos de filas:: producto A, producto B, producto C.  En las columnas las ventas por meses del año
import pandas as pd
import numpy as np
import random
import string

#df = pd.DataFrame(columns=['nombre', 'productoA', 'productoB', 'productoC'])
matrix = np.random.randint(1, 100, size=(12, 3))
df = pd.DataFrame(matrix, columns=['productoA', 'productoB', 'productoC'])
df
```

```python
#requiero ordenar las ventas en cada columna en orden ascendente
df.sort_values(by=['productoA', 'productoB', 'productoC'], ascending=True)
df
```

```python
#reemplaza el valor de 0 a 11 de las filas con los nombres de meses del año
df.index = ['enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre']
df


```

```python
#guarda el archivo como csv y muestra el resultado del csv
df.to_csv('ventas.csv')
df = pd.read_csv('ventas.csv')
df
```

```python
# Debes crear un gráfico para mostrar cómo han variado las ventas de cada producto mes a mes. Utiliza varios tipos de gráfico. Rotula los ejes y coloca como título: "Variación de ventas de productos por mes"

import pandas as pd
import matplotlib.pyplot as plt

df.plot(kind='bar', title='Variación de ventas de productos por mes')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.show()
```

```python
#salva la gráfica como pdf y déjala en este computador
import matplotlib.pyplot as plt

df.plot(kind='bar', title='Variación de ventas de productos por mes')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.savefig('variacion_ventas.pdf')
```

```python
#grafica para todo el año solo las ventas del producto A con título :: "Variación Anual Producto A", coloca lineas grid en la cuadrícula y coloca un label en cada punto de cada mes sobre la gráfica

df['productoA'].plot(kind='line', grid=True, title='Variación Anual Producto A')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.show()
```
